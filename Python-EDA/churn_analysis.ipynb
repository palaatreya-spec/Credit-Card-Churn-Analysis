# Credit Card Customer Churn Analysis

## Objective
The objective of this analysis is to identify the factors influencing customer churn in a credit card portfolio.

Understanding these drivers will help the business improve customer retention strategies and increase revenue.

Dataset Size: 10,127 customers
Features: Demographics, card usage, credit limit, transaction behavior.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_excel("../Excel/Excel_Analysis.xlsx", sheet_name="Dataset")
df.head()
```

```python
df.shape
```

## 1. What is the overall customer churn rate?

Understanding the proportion of customers leaving the service helps assess the severity of the attrition problem.

```python
plt.figure(figsize=(6,4))
sns.countplot(x="Attrition_Flag", data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

### Insight
Approximately **16% of customers have churned**, indicating a moderate attrition rate.

This suggests that the company should focus on identifying behavioral patterns among churned customers to design effective retention strategies.

## 2. Does card category influence churn?

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Card_Category", hue="Attrition_Flag", data=df)
plt.title("Churn by Card Type")
plt.show()
```

### Insight
Customers using **Blue cards show the highest churn rate** compared to premium card categories.

This suggests that lower-tier cardholders may have weaker engagement or fewer incentives to stay.

## 3. How does customer inactivity affect churn?

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Attrition_Flag", y="Months_Inactive_12_mon", data=df)
plt.title("Customer Inactivity vs Churn")
plt.show()
```

### Insight
Customers who remain inactive for longer periods are significantly more likely to churn.

This indicates that inactivity is a strong early warning signal for potential churn.

## 4. Does transaction behavior influence churn?

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Attrition_Flag", y="Total_Trans_Amt", data=df)
plt.title("Transaction Amount vs Churn")
plt.show()
```

### Insight
Customers with lower transaction amounts tend to churn more frequently.

Low transaction activity indicates low engagement with the credit card services.

## 5. Does credit utilization impact churn?

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Attrition_Flag", y="Avg_Utilization_Ratio", data=df)
plt.title("Credit Utilization vs Churn")
plt.show()
```

### Insight
Customers with very low credit utilization ratios are more likely to churn.

This suggests that customers who rarely use their credit card may not perceive sufficient value in maintaining the account.

## Key Findings

1. The overall churn rate is approximately **16%**.
2. **Blue card customers** show the highest churn rates.
3. **Customer inactivity** is a strong indicator of potential churn.
4. Lower **transaction activity** correlates with higher churn.
5. Customers with **low credit utilization** tend to disengage from the service.

These insights suggest that improving customer engagement and monitoring inactivity can help reduce churn.
